```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = [
    "DistilBERT (uncased)",
    "BERT (uncased)",
    "RoBERTa-base",
    "ALBERT-base"
]

# Metrics
accuracy = [93.8, 94.05, 93.95, 93.6]
f1_score = [93.79, 94.06, 93.97, 93.65]

# X-axis positions
x = np.arange(len(models))
width = 0.35  # width of bars

# Plot grouped bar chart
fig, ax = plt.subplots(figsize=(10,6))
rects1 = ax.bar(x - width/2, accuracy, width, label='Accuracy', color='skyblue')
rects2 = ax.bar(x + width/2, f1_score, width, label='F1 Score', color='salmon')

# Labels and title
ax.set_ylabel('Percentage (%)')
ax.set_title('Comparison of Emotion Classification Models')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=20)
ax.legend()

# Display values on top of bars
for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
models = [
    "martin-ha/toxic-comment-model",
    "Tiny-toxic-detector",
    "Stanford LSTM+BERT",
    "Naive Bayes + SVM"
]

# Corresponding accuracies
accuracy = [94, 90.26, 87.57, 85]

# X positions
x = np.arange(len(models))
width = 0.6

# Plot bar chart
plt.figure(figsize=(10,6))
bars = plt.bar(x, accuracy, width, color='skyblue')

# Labels and title
plt.ylabel('Accuracy (%)')
plt.title('Bias Model Accuracy Comparison')
plt.xticks(x, models, rotation=25)

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{height:.2f}%', ha='center')

plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
